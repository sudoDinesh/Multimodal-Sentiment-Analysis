```python
import pandas as pd

csv_path = r'/Users/dinesh/Downloads/mosei/CMU-MOSEI-20230514T151450Z-001/CMU-MOSEI/Labels/Data_Train_modified.csv'
df = pd.read_csv(csv_path)
```

```python
print(df.head())
```

```python

from pydub import AudioSegment
import os

audio_dir = r'/Users/dinesh/Downloads/mosei/CMU-MOSEI-20230514T151450Z-001/CMU-MOSEI/Audio_chunk'

def segment_audio(audio_path, start_time, end_time):
    audio = AudioSegment.from_wav(audio_path)
    segment = audio[start_time * 1000:end_time * 1000]  # convert to milliseconds
    return segment

for index, row in df.iterrows():
    video_id = row['video']
    start_time = row['start_time']
    end_time = row['end_time']
    audio_path = os.path.join(audio_dir, f"{video_id}.wav")
    if os.path.exists(audio_path):
        segment = segment_audio(audio_path, start_time, end_time)
        segment.export(f"{video_id}_{start_time}_{end_time}.wav", format="wav")
    else:
        print(f"No match for {video_id}")
```
